```python
import re

def process_document_structure(doc):
    structure = []
    current_section = None
    
    for para in doc.paragraphs:
        text = para.text.strip()
        
        # Определение уровня заголовка
        if re.match(r'^\d+\.', text):  # Раздел
            current_section = {'title': text, 'content': []}
            structure.append(current_section)
        elif re.match(r'^\d+\.\d+\.', text):  # Подраздел
            current_section['content'].append({'title': text, 'content': []})
        else:
            if current_section:
                current_section['content'].append(text)
    
    return structure


def smart_clean_text(text):
    # Удаляем только лишние пробелы между словами
    text = re.sub(r'\s+', ' ', text)
    # Сохраняем важные символы
    text = re.sub(r'[^\w\s.,;:—–()]', '', text)
    return text.strip()
```

```python
from docx import Document

# Загрузка документа
doc = Document('СП 60.13330.2020_ver5.docx')

process_document_structure(doc)

```
