### 1. Последовательности (Sequence )

Возрастающая последовательность:
$$ 
    x_{n+1} \geq x_{n}
$$

Убывающая последовательность:
$$ 
    x_{n+1} \leq x_{n}
$$

Ограниченная последовательность — это последовательность, все элементы которой лежат в некотором конечном интервале.
Монотонная последовательность - возрастающая или неубывающая (убывающая или невозрастающая). Изменения в одну сторону. 


````python
[(2**n + n**2) for n in range(15)]
````

````python
[(n - math.sqrt(n**2 - 3*n + 6)) for n in range(5)]
````

<h4> 1. Двойное Отрицание</h4>

$$
\neg(\neg A) \iff A
$$

<h4> 2. Отрицание Конъюнкции</h4>

$$
\neg(A \land B) \iff (\neg A \lor \neg B)
$$

<h4> 3. Отрицание Дизъюнкции</h4>

$$
\neg (A \lor B) \iff (A \land B)
$$
```python
A = [True, False]
B = [True, False]

for a in A:
    ai = a
    for b in B:
        bi = b
        print(f"A_{ai}, B_{bi}= {not(ai <= bi) == (ai and (not bi))}")
```

<h4> 4. Отрицание Импликации</h4>

$$
\neg (A \to B) \iff (A \land \neg B)
$$

```python
A = [True, False]
B = [True, False]

for a in A:
    ai = a
    for b in B:
        bi = b
        print(f"A_{ai}, B_{bi}= {not(ai <= bi) == (ai and (not bi))}")
```


<h4> 5. Раскрытие скобок с отрицанием к условию, содержащему квантор</h4>

- квантор меняется; 
- отрицание переносится на функцию;

$$
\neg(\forall x \in X | A(x)) \iff \exists x \in X |  \neg A(x)
$$

$$
\neg (\exists x \in X | A(x)) \iff \forall x \in X | \neg A(x)
$$

<h4> 6. Декартово произведение множеств (Cartesian product)</h4>

Декартовым произведением множеств X и Y называется множество X x Y, состоящиее из всех (упорядоченных ) пар (x, y) таких, что x in X, y in Y. 

X = {0, 1}

Y = {y0, y1, y2}

$$
X x Y = {(0, y0), (1, y0), (0, y1), (1, y1),(0, y2), (1, y2)}
$$
