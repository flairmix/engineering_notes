```python
people_1zone = 1778
people_2zone = 1161
people_3zone = 582
people_4zone = 230

people_sum = people_1zone + people_2zone + people_3zone + people_4zone

Q_1zone = 1.1155
Q_2zone = 0.7838
Q_3zone = 0.4554
Q_4zone = 0.2319

Q_sum = Q_1zone + Q_2zone + Q_3zone + Q_4zone

qhru = 6.5
q0_hr = 200
```

```python
people = people_sum-50
Q_Gcal_h = 0
result = {}

while Q_Gcal_h < Q_sum:

    flow_m3_day = people*70/1000
    flow_m3_hour = people*70/1000/24
    Q_mid_Gcal_h = flow_m3_hour * 60 / 1000
    NPhv = calculate_NPhv(qhru, q0_hr, people)
    alpha = calculate_alpha(NPhv)
    flow_max_hr_sum_m3_h = calculate_flow_max_hr(alpha, q0_hr)
    Q_Gcal_h = flow_max_hr_sum_m3_h * 60 * 1.15 / 1000

    result = {
        'people': people,
        'м3/сут': flow_m3_day,
        'м3/ч': flow_m3_hour,
        "NPhv": NPhv,
        "alpha": alpha,
        "flow_max_hr_sum_m3_h": flow_max_hr_sum_m3_h,
        "Q_mid_Gcal_h": Q_mid_Gcal_h,
        "Q_Gcal_h": Q_Gcal_h,
    }

    people += 1

for k, v in result.items():
    print (f"{k} - {v}")
```
