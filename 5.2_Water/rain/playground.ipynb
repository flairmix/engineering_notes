```python

def calculate_Q_r(Z_mid, A, F_r, t_r, n):
    """
    Расчёт расхода дождевых вод (Q_r) по формуле (Ж.1):
    Q_r = (Z_mid * A^(1.2) * F_r) / (t_r^(1.2*n - 0.1))
    
    Параметры:
    - Z_mid: среднее значение коэффициента покрова (из таблиц Ж.6 и Ж.7)
    - A: параметр, характеризующий расчётную интенсивность дождя (из формулы Ж.2)
    - F_r: расчётная площадь стока (га), ≤ 150 га
    - t_r: расчётная продолжительность дождя (с)
    - n: показатель степени (из таблицы Ж.1)
    
    Возвращает:
    - Q_r: расход дождевых вод в л/с
    """
    numerator = Z_mid * (A ** 1.2) * F_r
    denominator = t_r ** (1.2 * n - 0.1)
    Q_r = numerator / denominator
    return round(Q_r, 4)

def calculate_A(q_20:float = 91,
                n:float = 0.71,
                P:float=1,
                m_r:float = 150,
                y:float = 1.54,
                ):
    """
    Расчёт параметра A по формуле (Ж.2):
    A = q_20 * 20^n * (1 + lg(P) / lg(m_r))**y
    
    Параметры:
    - q_20: интенсивность дождя (20 мин, P=1 год) — из рисунка Ж.1
    - n: показатель степени (из таблицы Ж.1)
    - m_r: среднее количество дождей за год (из таблицы Ж.1)
    - P: период однократного превышения расчётной интенсивности дождя (годы)
    - y: показатель степени
    
    Возвращает:
    - A: параметр для расчёта расхода дождевых вод
    """
    lg_P = math.log10(P)
    lg_m_r = math.log10(m_r)

    A = q_20 * (20 ** n) * (1 + (lg_P / lg_m_r))**y
    return A

```

```python
def Z_mid_calculate_weighted_average(coefficients_dict, areas_dict):
    """
    Рассчитывает средневзвешенное значение коэффициентов по площадям.
    
    Параметры:
    -----------
    coefficients_dict : dict
        Словарь с коэффициентами для разных типов покрытий.
        Пример: {"Waterproof surfaces": 0.33, "Paving stones...": 0.224, ...}
    
    areas_dict : dict
        Словарь с площадями (в одинаковых единицах) для каждого типа покрытия.
        Ключи должны совпадать с ключами coefficients_dict.
        Пример: {"Waterproof surfaces": 100, "Paving stones...": 50, ...}
    
    Возвращает:
    ------------
    float
        Средневзвешенное значение коэффициента.
        Если суммарная площадь = 0, возвращает 0.0.
    
    Raises:
    -------
    KeyError
        Если в areas_dict есть ключи, отсутствующие в coefficients_dict.
    ValueError
        Если площади отрицательны.
    """
    total_weighted_sum = 0.0
    total_area = 0.0
    
    for cover_type, area in areas_dict.items():
        # Проверка наличия типа покрытия в словаре коэффициентов
        if cover_type not in coefficients_dict:
            raise KeyError(f"Тип покрытия '{cover_type}' отсутствует в словаре коэффициентов.")
        
        # Проверка неотрицательности площади
        if area < 0:
            raise ValueError(f"Площадь для '{cover_type}' отрицательна: {area}")
        
        coefficient = coefficients_dict[cover_type]
        total_weighted_sum += coefficient * area
        total_area += area
    
    # Если общая площадь нулевая, возвращаем 0.0 (избегаем деления на ноль)
    if total_area == 0:
        return 0.0
    
    weighted_average = total_weighted_sum / total_area
    return round(weighted_average, 3)


```

```python
# city_rain_parameters # result['P_1'] = 91.0
q_20 = 91

# city_rain_parameters # result['n_P>1'] = 0.71
n = 0.71 

P = 1       # период превышения (годы)

# city_rain_parameters # result['mr'] = 150
m_r = 150     # среднее количество дождей за год, из таблицы Ж.1

# city_rain_parameters # result['mr'] = 1.54
y = 1.54

Z_mid = Z_mid_calculate_weighted_average(Zi_cover_coefficients, areas)

F_r = 1     # расчётная площадь стока (га) (<150 га)

# tcon - продолжительность протекания дождевых вод до уличного лотка или при наличии дождеприемников 
# в пределах квартала до уличного коллектора (время поверхностной концентрации), мин, определяется согласно Ж.6;
# Ж.6 Время поверхностной концентрации дождевого стока tcon следует рассчитывать или, 
# при отсутствии внутриквартальных закрытых дождевых сетей в поселениях и городских округах, принимать 
# 5 - 10 мин, а при их наличии - 3 - 5 мин. 
# При расчете внутриквартальной канализационной сети время поверхностной концентрации следует принимать 2 - 3 мин.
t_con = 5

# Продолжительность протекания дождевых вод по уличным лоткам tcan следует определять по формуле Ж.4
# где lcan - длина участков лотков, м;
# lcan - расчетная скорость течения на участке, м/с.
t_can = 0
l_v_can = {
    0 : 1.0,
}
for l, v in l_v_can.items():
    t_can += l/v
t_can *= 0.021
 
# Продолжительность протекания дождевых вод по трубам до рассчитываемого сечения tр, мин, следует определять по формуле Ж.5
# где lр - длина расчетных участков коллектора, м;
# vр - расчетная скорость течения на участке, м/с.
t_p = 0
l_v_p = {
    100 : 1.0,
}
for l, v in l_v_p.items():
    t_p += l/v
t_p *= 0.017

# Ж.5 Расчетную продолжительность протекания дождевых вод по поверхности и трубам до расчетного участка (створа) tr, мин, следует определять по формул
t_r = t_con + t_can + t_p

print(f"t_r - {t_r} мин")

# Сначала вычисляем параметр A
A = calculate_A(q_20=91, 
                n=0.71, 
                P=1,
                m_r=150, 
                y=1.54
                )
print(f"Параметр A = {A:.2f}")

# Затем рассчитываем расход дождевых вод Q_r
Q_r = calculate_Q_r(Z_mid, A, F_r, t_r, n)
print(f"Расход дождевых вод Q_r = {Q_r:.2f} л/с")
```
