```python
from typing import List, Dict, Optional

class Parking_UG:
    R_fl = (2.1, 4.3, 8.6, 14.2)

    def __init__(
        self,
        name: str = 'parking_UG_',
        levels: Optional[List[Dict]] = None
    ):
        self.name = name
        self.levels = levels if levels is not None else []
        self.area_m2_IV = 0
        self.perimeters_m = self.zone_perimeters()
        self.zones_area_m2 = [p*2 for p in self.perimeters_m]
        self.zones_area_m2[3] = self.area_m2_IV
        self.zones_Qhl_W_K = [int(area / R) for area, R in zip(self.zones_area_m2, self.R_fl)]


    def zone_perimeters(self) -> List[float]:
        perimeters = [0.0, 0.0, 0.0, 0.0]  

        if len(self.levels) == 1:
            level = self.levels[0]  
            height_m = level['height_m'] 
            perimeter = level['perimeter_m']
            self.area_m2_IV = level['area_m2']

            i = 0
            while (height_m >= 2 and i < 3):
                perimeters[i] = perimeter
                i += 1
                height_m -= 2
            
            if height_m < 2 and height_m != 0:
                perimeters[i] = perimeter
                i += 1
                perimeter = ((perimeter / 4) - ((2-height_m*2))) * 4
                self.area_m2_IV -= perimeter * (2-height_m)
                height_m = 0

            while i < 4 and height_m == 0:
                perimeters[i] = perimeter
                self.area_m2_IV -= ((perimeter / 4) - 4) * 4
                i += 1

        return perimeters
    

```
