```python
import numpy as np
from scipy import integrate, optimize
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 * np.sin(x)
```

```python

x = np.linspace(-50, 50, 400)  # 400   -10  10
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = x^2 \sin(x)$', color='blue')
plt.title(' ')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
min_point = optimize.minimize_scalar(f, bounds=(0, 50), method='bounded')
print(f"  x = {min_point.x:.6f}, f(x) = {min_point.fun:.6f}")
```

```python
dy_dx = np.gradient(y, x)  # 

plt.figure(figsize=(10, 6))
plt.plot(x, dy_dx, label="$f'(x)$", color='red')
plt.title(' ')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def f(x):
    return x**2 * np.sin(x)

def df_analytical(x):
    return 2*x*np.sin(x) + x**2*np.cos(x)  #  f'(x)  !
```

```python
x = np.linspace(0, 5, 100)
y = f(x)

#   ()
dy_dx_numerical = np.gradient(y, x)


#   ()
dy_dx_analytical = df_analytical(x)

# 
plt.plot(x, dy_dx_numerical, label=' (np.gradient)', linestyle='--')
plt.plot(x, dy_dx_analytical, label=' ()', linewidth=2)
plt.legend()
plt.show()
```

```python
result, error = integrate.quad(f, 0, 1)
print(f": {result:.6f}, : {error:.2e}")
```

```python
result, error = integrate.quad(f, 0, 10)
print(f": {result:.6f}, : {error:.2e}")
```

```python
F = integrate.cumulative_trapezoid(y, x, initial=0)  #  0

plt.figure(figsize=(10, 6))
plt.plot(x, F, label="$F(x) = \int_{x_0}^{x} f(t)\,dt$", color='purple')
plt.title(' ')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
min_point = optimize.minimize_scalar(f, bounds=(-25, 25), method='bounded')
print(f"  x = {min_point.x:.6f}, f(x) = {min_point.fun:.6f}")
```
