```python
import numpy as np
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

A1 = np.array([[1,2,3], [4,5,6]])
B1 = np.array([[1,0,1], [0,1,-1]])
```

```python
#Сложение/вычитание  
A1 + B1
```

```python
#Умножение на скаляр
v1 * 2
```

```python
#Скалярное произведение векторов
# [1,2,3] @ [4,5,6] = 1*4 + 2*5 + 3*6 = 4+10+18 = 32
print(v1 @ v2)
```

```python
#Матричное умножение

A2 = np.array([[1,2], [3,4], [5,6]])
B2 = np.array([[1,0,1], [0,1,-1]])

print(A2)
print(B2)

print(A2 @ B2)
```
