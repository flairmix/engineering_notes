```python
import folium
```

```python

def folium_Map_generate(coords:dict, 
                        name_output:str = "map"):
    m = folium.Map(location=[60, 100], zoom_start=3)

    for city, (lat, lon) in coords.items():
        folium.Marker([lat, lon], popup=city).add_to(m)

    m.save(f'{name_output}.html')
```
